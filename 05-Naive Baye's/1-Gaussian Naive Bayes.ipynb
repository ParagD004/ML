```python
# Import all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Set style for better plots
plt.style.use('default')
sns.set_palette("viridis")
```

```python
class GaussianNaiveBayes:
    """
    Gaussian Naive Bayes classifier implementation
    """
    
    def __init__(self):
        self.classes_ = None
        self.class_priors_ = None
        self.means_ = None
        self.variances_ = None
    
    def fit(self, X, y):
        """
        Fit Gaussian Naive Bayes to training data
        """
        self.classes_ = np.unique(y)
        n_classes = len(self.classes_)
        n_features = X.shape[1]
        
        # Initialize arrays for means and variances
        self.means_ = np.zeros((n_classes, n_features))
        self.variances_ = np.zeros((n_classes, n_features))
        self.class_priors_ = np.zeros(n_classes)
        
        # Calculate priors, means, and variances for each class
        for i, c in enumerate(self.classes_):
            X_c = X[y == c]
            self.class_priors_[i] = X_c.shape[0] / X.shape[0]
            self.means_[i, :] = X_c.mean(axis=0)
            self.variances_[i, :] = X_c.var(axis=0)
    
    def _calculate_likelihood(self, x, mean, var):
        """
        Calculate Gaussian likelihood
        """
        # Add small epsilon to avoid division by zero
        eps = 1e-9
        coeff = 1.0 / np.sqrt(2 * np.pi * var + eps)
        exponent = np.exp(-(x - mean) ** 2 / (2 * var + eps))
        return coeff * exponent
    
    def predict(self, X):
        """
        Predict class labels for samples in X
        """
        n_samples = X.shape[0]
        n_classes = len(self.classes_)
        predictions = np.zeros(n_samples, dtype=self.classes_.dtype)
        
        for i in range(n_samples):
            posteriors = []
            
            # Calculate posterior probability for each class
            for j in range(n_classes):
                prior = np.log(self.class_priors_[j])
                likelihood = np.sum(np.log(self._calculate_likelihood(
                    X[i, :], self.means_[j, :], self.variances_[j, :])))
                posterior = prior + likelihood
                posteriors.append(posterior)
            
            # Predict class with highest posterior probability
            predictions[i] = self.classes_[np.argmax(posteriors)]
        
        return predictions
    
    def predict_proba(self, X):
        """
        Predict probability estimates for samples in X
        """
        n_samples = X.shape[0]
        n_classes = len(self.classes_)
        probabilities = np.zeros((n_samples, n_classes))
        
        for i in range(n_samples):
            class_probs = []
            
            # Calculate unnormalized probabilities for each class
            for j in range(n_classes):
                prior = np.log(self.class_priors_[j])
                likelihood = np.sum(np.log(self._calculate_likelihood(
                    X[i, :], self.means_[j, :], self.variances_[j, :])))
                class_probs.append(prior + likelihood)
            
            # Convert to probabilities using softmax
            class_probs = np.array(class_probs)
            # Subtract max for numerical stability
            class_probs = class_probs - np.max(class_probs)
            exp_probs = np.exp(class_probs)
            probabilities[i, :] = exp_probs / np.sum(exp_probs)
        
        return probabilities
```

```python
# Load and prepare the Iris dataset
print("=== Loading Iris Dataset ===")
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Create DataFrame for easier visualization
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y
df['species'] = [target_names[i] for i in y]

print("Dataset Information:")
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Feature names: {feature_names}")
print(f"Target names: {list(target_names)}")
print(f"\nClass distribution:")
print(df['species'].value_counts())

# Display first few rows
print("\nFirst 5 rows of the dataset:")
display(df.head())
```

```python
# Exploratory Data Analysis - Pairplot
print("=== Pairplot of Iris Dataset ===")
plt.figure(figsize=(12, 8))
sns.pairplot(df, hue='species', palette='viridis', diag_kind='hist')
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02)
plt.show()
```

```python
# Exploratory Data Analysis - Feature Distributions by Class
print("=== Feature Distributions by Species ===")
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

for i, feature in enumerate(feature_names):
    for species in target_names:
        species_data = df[df['species'] == species][feature]
        axes[i].hist(species_data, alpha=0.7, label=species, bins=15)
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
# Split the data into training and testing sets
print("=== Splitting Data into Training and Test Sets ===")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
print(f"Training class distribution: {np.bincount(y_train)}")
print(f"Test class distribution: {np.bincount(y_test)}")
```

```python
# Train Gaussian Naive Bayes model
print("=== Training Gaussian Naive Bayes Model ===")
gnb = GaussianNaiveBayes()
gnb.fit(X_train, y_train)

print("Model trained successfully!")
print(f"Classes: {gnb.classes_}")
print(f"Class priors: {gnb.class_priors_}")
```

```python
# Display model parameters
print("=== Model Parameters ===")
for i, class_name in enumerate(target_names):
    print(f"\n{class_name}:")
    print(f"  Prior probability: {gnb.class_priors_[i]:.4f}")
    print(f"  Means: {[f'{m:.4f}' for m in gnb.means_[i]]}")
    print(f"  Variances: {[f'{v:.4f}' for v in gnb.variances_[i]]}")
```

```python
# Make predictions and evaluate model
print("=== Making Predictions and Evaluating Model ===")
y_pred = gnb.predict(X_test)
y_proba = gnb.predict_proba(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")

# Display classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
# Plot Confusion Matrix
print("=== Confusion Matrix ===")
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Display prediction examples with probabilities
print("=== Prediction Examples ===")
print("Index\tTrue\t\tPredicted\tProbabilities")
print("-" * 65)

for i in range(min(15, len(y_test))):
    true_label = target_names[y_test[i]]
    pred_label = target_names[y_pred[i]]
    probs = [f"{p:.3f}" for p in y_proba[i]]
    
    # Add marker if prediction is incorrect
    marker = " ✓" if true_label == pred_label else " ✗"
    
    print(f"{i}\t{true_label:15}{pred_label:15}{probs}{marker}")
```
